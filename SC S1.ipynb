```python
import numpy as np

X = np.array ([-1, -0.9, -0.8, -0.7, -0.6])
F = np.array ([-0.63212, -0.49343, -0.35067, -0.20341, -0.05119])

h = X[1] - X[0]

#Second derivatife (f'')
x = -1
f_second_derivative = (2*F[0] - 5*F[1] + 4*F[2] - 1*F[3]) / (h**2)

#Fourt derivatife (f'''')
f_fourth_derivative = (3*F[0] - 14*F[1] + 26*F[2] - 24*F[3] + 11*F[4]) / (h**4)


print ("Second derivative at x = -1:", f_second_derivative)
print ("Fourt derivative at x = -1:", f_fourth_derivative)
```
