```python
x_values = [-1, -0.9, -0.8, -0.7, -0.6]
f_values = [-0.63212, -0.49343, -0.35067, -0.20341, -0.05119]

# STEP SIZE
h = x_values[1] - f_values[0]

f_second_derivative = (f_values[0] - 2*f_values[1] + f_values[2]) / (h**2)
f_fourth_derivative = (f_values[0] - 4*f_values[1] + 6*f_values[2] - 4*f_values[3] + f_values[4]) / (h**4)

# Print the results
print("Second derivative (f''(-1)):", f_second_derivative)
print("Fourth derivative (f^4(-1)):", f_fourth_derivative)
```
