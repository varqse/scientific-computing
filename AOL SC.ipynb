```python
import numpy as np

years = np.array([1981, 1983, 1985, 1987, 1989, 1991, 1993, 1995, 1997, 1999])
temperatures = np.array([14.1999, 14.2411, 14.0342, 14.2696, 14.197, 14.3055, 14.1853, 14.3577, 14.4187, 14.3438])

even_years = np.arange(1982, 2000, 2)

linear_temperatures = np.interp(even_years, years[::2], temperatures[::2])

quadratic_coefficients = np.polyfit(years[::2], temperatures[::2], 2)
quadratic_temperatures = np.polyval(quadratic_coefficients, even_years)

cubic_coefficients = np.polyfit(years[:4], temperatures[:4], 3)
cubic_temperatures = np.polyval(cubic_coefficients, even_years)

print("Tahun Genap   | Interpolasi Linear | Interpolasi Kuadrat | Interpolasi Kubik")
print("-----------------------------------------------")
for year, linear_temp, quadratic_temp, cubic_temp in zip(even_years, linear_temperatures, quadratic_temperatures, cubic_temperatures):
    print(f"{year:<13}| {linear_temp:.4f}          | {quadratic_temp:.4f}         | {cubic_temp:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

Year = np.array([1981, 1983, 1985, 1987, 1989, 1991, 1993, 1995, 1997, 1999])
Temperature = np.array([14.1999, 14.2411, 14.0342, 14.2696, 14.1970, 14.3055, 14.1853, 14.3577, 14.4187, 14.3438])
x_regres2 = np.arange(1981, 2000)
y_regres2 = -9.93467757575738 + 0.01215576 * x_regres2

plt.plot(Year, Temperature, 'o-', label='Data')
plt.plot(x_regres2, y_regres2, 'o--', label='Regression Line')
plt.xlabel('Year')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.show()
```

```python
import sympy as sp

x = sp.Symbol('x')

sin_expansion = sp.series(sp.sin(x), x, 0, 5).removeO()
print("Ekspansi Taylor urutan keempat untuk sin(x):")
print(sin_expansion)

cos_expansion = sp.series(sp.cos(x), x, 0, 5).removeO()
print("Ekspansi Taylor urutan keempat untuk cos(x):")
print(cos_expansion)

product_expansion = sp.expand(sin_expansion * cos_expansion)
print("Perluasan Taylor urutan keempat untuk sin(x)cos(x) sekitar 0:")
print(product_expansion)

x_value = sp.pi/2

sin_value = sin_expansion.subs(x, x_value)
cos_value = cos_expansion.subs(x, x_value)
product_value_1 = sin_value * cos_value

product_value_2 = product_expansion.subs(x, x_value)

print("\nHasil Evaluasi untuk x = π/2:")
print("Metode 1 (Mengalikan perluasan terpisah):", product_value_1.evalf())
print("Metode 2 (Menghitung ekspansi untuk produk):", product_value_2.evalf())
```

```python
import sympy as sp

x = sp.Symbol('x')
x_value = sp.pi/4

cos_expansion = sp.series(sp.cos(x), x, 0, 5).removeO()

cos_approximation = cos_expansion.subs(x, x_value)
approximation_value = cos_approximation.evalf()

error_bound = abs(cos_expansion.args[-1].subs(x, x_value))
print("Perkiraan nilai cos(π/4):", approximation_value)
print("Batas kesalahan pemotongan:", error_bound.evalf())
```

```python
import math
def f(x):
    return x**3 - 0.3*x**2 - 8.56*x + 8.448
def riemann_integral():
    a = 0
    b = 2 * math.pi
    n = 20
    delta_x = (b - a) / n
    integral_sum = 0

    for i in range(n):
        x_i = a + i * delta_x
        integral_sum += f(x_i) * delta_x

    return integral_sum
def trapezoidal_rule():
    a = 0
    b = 2 * math.pi
    n = 20
    delta_x = (b - a) / n
    integral_sum = 0

    for i in range(n):
        x_i = a + i * delta_x
        x_i_plus_1 = a + (i + 1) * delta_x
        integral_sum += (f(x_i) + f(x_i_plus_1)) * delta_x / 2

    return integral_sum
def simpsons_rule():
    a = 0
    b = 2 * math.pi
    n = 20
    delta_x = (b - a) / n
    integral_sum = 0

    for i in range(n):
        x_i = a + i * delta_x
        x_i_plus_half = a + (i + 0.5) * delta_x
        x_i_plus_1 = a + (i + 1) * delta_x
        integral_sum += (f(x_i) + 4 * f(x_i_plus_half) + f(x_i_plus_1)) * delta_x / 6

    return integral_sum

integral_riemann = riemann_integral()
integral_trapezoidal = trapezoidal_rule()
integral_simpsons = simpsons_rule()

print("Kira-kira integral menggunakan metode Riemann Integral:", integral_riemann)
print("Kira-kira integral menggunakan metode Aturan Trapesium:", integral_trapezoidal)
print("Kira-kira integral menggunakan metode Aturan Simpson:", integral_simpsons)
```
